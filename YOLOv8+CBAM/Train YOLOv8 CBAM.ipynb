```python
import csv
import glob
import time
import torch
import ultralytics
import pandas as pd
from ultralytics import YOLO

%matplotlib inline
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
namaGes = "YOLOv8m-Seg.pt"
ultralytics.checks(device='0')
```

# Load Model Config

```python
modelConfig = {}
for model in sorted(glob.glob(r"Config Model/*/*.yaml")):
    modelName = model.split("\\")[1]
    modelConfig[modelName] = model
    print(f"Model '{modelName}' Loaded Sucessfully from {model}")
```

# Training

```python
def trainingGoBrrr(modelName, namaRun):
    
    torch.cuda.empty_cache()
    model = YOLO(modelName).load("yolov8m-seg.pt")
    model.train(
        data="D:/CODE/06 Cardiomegali/P2/code/Master YOLOv8 (CBAM, ResBlockCBAM, ResNeXtBlockCBAM)/data.yaml",
        epochs=100, 
        batch=4,
        imgsz=640,
        patience=0,
        optimizer='SGD',
        name=namaRun
    )

    return model
```

```python
modelDict = {}

for model in modelConfig.keys():
    print("\n", "="*80)
    print(f"Sekarang Training Model {model}")
    
    torch.cuda.empty_cache()
    modelDict[model] = trainingGoBrrr(modelConfig[model], model)
    time.sleep(20)
```
